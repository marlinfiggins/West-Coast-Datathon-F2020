```python
movies = pd.read_csv("../data/movies.csv")
movies
```

```python
category  = movies.genres.str.split("|").to_list()
category = [j for sub in category for j in sub] 
category = np.unique(category)
category
```

```python
for i in range(len(category)):
    movies[category[i]] = movies.genres.str.contains(category[i])*1
    movies[category[i]]
```
